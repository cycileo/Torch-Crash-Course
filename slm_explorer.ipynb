## 1. The Model and The Data
The model powering this tool is a **Mini-GPT**, a simplified, scaled-down version of the Decoder-only Transformer architecture that powers massive models like ChatGPT. 

### Character-Level Tokenization
Modern language models process text in "chunks" (sub-words), but to make the mechanics as transparent as possible, our model operates at the **character level**. It treats every single letter, space, and punctuation mark as an individual token. 
* The vocabulary consists of 39 unique lowercase characters.
* The model was trained entirely on the **Tiny Shakespeare** dataset (roughly 1 million characters of Shakespearean plays). 

### The Architecture
Under the hood, the model relies on the core innovations of the Transformer:
1. **Embeddings:** Converts characters into high-dimensional vectors, injecting mathematical positional data so the model understands the sequence order.
2. **Masked Self-Attention:** The model looks at the context of previous characters to figure out what should come next. The "Causal Mask" ensures it can only look at the *past*, never the *future*.
3. **Linear Head:** A final layer that outputs raw scores (logits) for all 65 possible next characters.

## 2. The Autoregressive Objective
A language model is essentially a conditional probability engine. Its entire goal is to estimate the probability distribution of the next token ($x_t$), given the context of all previous tokens ($x_{<t}$):

$$P(x_t \mid x_1, \dots, x_{t-1})$$

Because it uses its own past predictions as the context for future predictions, the generation process is called **autoregressive**. 

## 3. How the Explorer Works (Frontend & Backend)
This notebook runs a full-stack web application directly inside your Jupyter kernel:
* **The Backend (PyTorch + Flask):** A background server holds the trained neural network in memory. When you interact with the UI, the server intercepts your sequence, crops it to the model's maximum context length, and runs a live forward pass on the GPU/CPU. It extracts the logits for the final token, calculates the Top-10 probabilities, and sends them back.
* **The Frontend (HTML/JS):** A reactive user interface that visualizes these probabilities in real-time without needing to constantly reload the page.

## 4. How to Experiment
Run the cell below to launch the UI, then try the following:

* **Hover to Inspect:** Place your mouse over any character in the generated text. The UI will highlight the exact context the model used and display the Top-10 probability distribution for what it thought should come next.
* **Click to Branch:** Language generation is a tree of possibilities. Click on any character to slice the sequence. You can then generate a completely different timeline from that exact point forward.
* **Temperature Tuning:** The Temperature slider controls the "creativity" or entropy of the model. Mathematically, it scales the raw logits ($z_i$) before the Softmax function turns them into probabilities:
  
  $$p_i = \frac{\exp(z_i / T)}{\sum \exp(z_j / T)}$$
  
  * **$T = 1.0$ (Default):** The true distribution learned by the model.
  * **$T < 1.0$ (Greedy):** As temperature approaches zero, the model becomes hyper-confident, heavily favoring the most likely token and ignoring the rest.
  * **$T > 1.0$ (Exploratory):** The probability distribution flattens. Lower-probability characters get a boost, resulting in more diverse, unpredictable, and sometimes nonsensical text.

You will immediately notice a major difference when running GPT-2: it no longer predicts single letters. 

Modern models use **Sub-word Tokenization** (like Byte-Pair Encoding). Instead of a vocabulary of 65 characters, GPT-2 has a vocabulary of over 50,000 tokens representing common words, prefixes, and syllables (e.g., " the", "ing", "apple"). This allows the model to pack more meaning into a shorter sequence and grasp complex semantics much faster.
